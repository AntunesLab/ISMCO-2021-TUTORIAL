# Get the antigen (protein) sequence from Uniprot
Uniprot is the reference database for curated protein sequences. The mission of UniProt is to provide the scientific community with a comprehensive, high-quality and freely accessible resource of protein sequence and functional information.

To access Uniprot, please click [here](https://www.uniprot.org/).
