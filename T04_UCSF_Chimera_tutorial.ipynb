## A) Loading and moving PDB structures

The UCSF Chimera will open the structure with default colors for the structure in a black background. Use your mouse to move the structure and find a position that suits you better. Here are some basic commands to work with your structure:



   - *Left mouse button*: controls rotation.
   - *Middle mouse button*: controls XY translation (panning).
   - *Right mouse button*: performs scaling (zooming).
   - *Left mouse button + Ctrl key*: select atoms, bonds or residues (to unselect, keep the button and Ctrl key pressed and click in the background) 
   - *Right mouse button + Ctrl key*: select a residue as pivot.
   - *Left mouse button + Ctrl key + Shift key*: select multiple atoms, bonds or residues (to unselect, keep the button and Ctrl key pressed and click in the background) 
   
   *Tip 1: You can also drag out a selection area with Ctrl + Left button (sweep out an area before releasing).*
   
   *Tip 2: The arrow keys can be used to broaden (), narrow (), or invert () a selection.* 
   
   *Tip 3: If you move the mouse cursor over a residue, you will see the following information displayed:* <br>
           Name of the residue + Number of the residue + Chain


## B) Selection and Actions

**1. Selection**: Chimera actions will normally be applied to the current selection. Let's see some of the available selections:<br>
<br>





 - *Chain*: select a specific chain, based on PDB file.
 - *Chemistry*: select parts of the molecule according to chemical features
 - *Residue*: select residues according to its category (e.g. aromatic, charged, polar), all nonstandard molecules (e.g. SO4, HOH) or one of 20 standard amino acids.
 - *Structure*: select a specific part of the molecule (e.g. ions, secondary structure, backbone)
 - *Sequence...*: look for a determined sequence and select it.
 - *Atom Specifier...*: select a specific atom
 - *Zone...*: select all atom/bons inside or outside a specific range from currently selected atoms.
 - *Clear selection*: remove all selections.
 - *Name Selection...*: after select some specific part, you can save that selection to be used later.

**2. Action**: You cant apply different actions to your selecion (PS: If you have nothing selected, the action normally applies to the whole molecule)<br>
<br>




- *Atoms/Bonds*: Controls the display and representation of atoms and bonds.
- *Ribbon*: Controls the display and representation of ribbons.
- *Surface*: Controls the display and representation of molecular surfaces.
- *Color*: Colors selected objects. Color target can be limited to object types in the all options dialog.
- *Label*: Labels selected atoms. The residue submenu labels residues containing the selected atoms.
- *Focus*: Focuses the view on the selected atom(s), zooming and translating if necessary.
- *Set Pivot*: Sets the center of rotation based on the selected atom(s) without adjusting the view.
- *Inspect*: Launches the Selection Inspector; same as clicking on the magnifying glass icon near the bottom right corner of the window.
- *Write List*: Writes a list of the currently selected objects to a parsable text file.
- *Write PDB*: Writes the coordinates of the currently selected atoms to a PDB file.

## C) Removing chains and saving new PDBs

<div style="text-align: justify"> Sometimes, we want to work with a specific part of a molecule. Using our previous example (5BRZ), suppose you are interested only in the pMHC structure. How can you remove a part of the molecule and save it in a different file? For that, follow these instructions:</div>






1. Select chain D (Select > Chain);
2. Delete the atoms/bonds (Actions > Atoms/Bonds > Delete);
3. Select chain E (Select > Chain);
4. Delete the atoms/bonds (Actions > Atoms/Bonds > Delete);
5. Save the resulting structure to a new file (File > Save PDB ...) as 5brz_pMHC.pdb.




6. Close the session (**Fille  Close session**)
7. Repeat the same process for the structure 5BS0. At the end, save as 5bs0_pMHC.pdb.

After you finished, close the session again (**Fille  Close session**).

## F) Command line

Chimera allows you to control the selection, actions and tools through command line. Here, we will see how to make different selections using command line tool. For that, open the command line panel through 'Tools' menu.<br>



Here are some of the symbols you can use to select different parts of your molecule.
 - "#"	--> model number
 - ":"	--> residue
 - ":." --> chain ID
 - "@"	--> atom name
