# <div style="text-align: center">PDB Tutorial</div>

<div style="text-align: justify">The PDB (Protein Data Bank) is a database for three-dimensional data from biomolecules, specially large molecules such as proteins and nucleic acids. The data is obtained from a variety of methods, including x-ray crystallography, Nuclear Magnetic Resonance (NMR), cryo-electron microscopy (Cryo-EM), and others. The data is freely accessible and overseen by Worldwide Protein Data Bank (wwPDB) </div>

To access PDB, please click [here](https://www.rcsb.org/).
